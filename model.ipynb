```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Assuming your time series data is stored in a Pandas DataFrame or Series called 'data'
# 'data' should have a column representing the time index and another column representing the values

# Calculate autocorrelation using statsmodels
acf_values = sm.tsa.acf(data.sales, nlags=100, fft=True)

# Plot the autocorrelation values
plt.figure(figsize=(10, 4))
plt.stem(range(len(acf_values)), acf_values)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function')
plt.show()
```

```python
lag = 7
```

```python
features = data[['timestamp', 'sales']]
features
```

```python
mean_timestamp = features.timestamp.mean()
std_timestamp = features.timestamp.std()
```

```python
standard_timestamp = features['timestamp'].apply(lambda x: (x-mean_timestamp)/std_timestamp)
standard_timestamp
```

```python
mean_sales = data['sales'].mean()
std_sales = data['sales'].std()
```

```python
standard_sales = data['sales'].apply(lambda x:(x-mean_sales)/std_sales)
standard_sales
```

```python
import pickle
pickle.dump([mean_sales, std_sales, mean_timestamp, std_timestamp], open('standard_vals.pkl','wb'))
```

```python
features.timestamp = standard_timestamp
features.sales = standard_sales
features
```

```python
import numpy as np
from scipy import stats

def detect_outliers_zscore(data, threshold=3):
    z_scores = np.abs(stats.zscore(data))
    outlier_indices = np.where(z_scores > threshold)[0]
    return outlier_indices

outlier_indices = detect_outliers_zscore(features['sales'], threshold=2.5)
print("Outlier indices:", outlier_indices)
```

```python
features['sales'][outlier_indices] = 0
```

```python
targets = features[['sales']].iloc[lag:]
targets
```

```python
inputs = np.asarray([features[['timestamp','sales']].iloc[x:x+lag].values for x in range(0,features.shape[0]-lag)])
inputs.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test,  y_train, y_test = train_test_split(inputs, targets, test_size=.2,random_state=1234)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
```

```python
model = keras.Sequential([
    layers.InputLayer((7,2)),
    layers.Dense(16, activation='elu'),
    layers.LSTM(32,dropout=.5, recurrent_dropout=.5, return_sequences=True),
    layers.LSTM(64, dropout=.5, recurrent_dropout=.5),
    layers.Dense(128, activation='elu'),
    layers.Dropout(.2),
    layers.Dense(64, activation='elu'),
    layers.Dropout(.2),
    layers.Dense(32, activation='elu'),
    layers.Dense(1)
])
model.summary()
```

```python
model.compile('adam',loss='mse',metrics=['mae','mape'])
```

```python
history = model.fit(x_train, y_train, 1000,500,1)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
preds = (predictions*std_sales) + mean_sales
actual = (y_test * std_sales) + mean_sales
```

```python
plt.figure(figsize=(20,3))
plt.plot(list(range(len(predictions))),preds,c='red', label='preds')
plt.plot(list(range(len(predictions))),actual, alpha=0.5, label='actual')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(actual, preds), r2_score(y_test, predictions)
```

```python
num_predictions = 20
temp = features.iloc[-7:]
date_range = pd.date_range(temp.index[-1],periods=num_predictions+1,freq='D')
for i in range(num_predictions):
    time_diff = temp['timestamp'][2] - temp['timestamp'][1]
    prediction = float(model.predict(np.array([temp.iloc[-7:].values]), verbose=0).squeeze())
    pred = pd.DataFrame({'timestamp':float(temp.timestamp[-1]+time_diff),'sales':prediction}, index = [date_range[i+1]])
    temp = pd.concat([temp,pred],axis=0)
```
